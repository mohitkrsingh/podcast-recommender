```python
# Load the dataset
df = pd.read_csv('df_popular_podcasts.csv')

# Drop unnecessary columns
df.drop(columns=[
    'Genre IDs',
    'Episode Count',
    'Episode Durations',
    'Feed URL'
], inplace=True)

# Check final columns
print("Columns after cleanup:\n", df.columns.tolist())
df.head()
```

```python
import spacy
import re
from tqdm import tqdm  # to show progress bars

nlp = spacy.load("en_core_web_sm")
```

```python
def clean_description(text):
    # 1. Lowercase everything
    text = text.lower()

    # 2. Remove all characters that are NOT alphabets
    text = re.sub(r'[^a-z\s]', '', text)

    # 3. Use spaCy to process the text
    doc = nlp(text)

    # 4. Remove stopwords and punctuation; lemmatize the remaining words
    tokens = [token.lemma_ for token in doc if not token.is_stop and token.is_alpha]

    # 5. Join cleaned tokens back into a string
    return ' '.join(tokens)
```

```python
tqdm.pandas()  # adds a progress bar to apply()
df['cleaned_description'] = df['Description'].progress_apply(clean_description)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Initialize the vectorizer
tfidf = TfidfVectorizer(stop_words='english')

# Fit and transform the cleaned descriptions
tfidf_matrix = tfidf.fit_transform(df['cleaned_description'])

print(tfidf_matrix.shape)  # Should show (num_podcasts, num_unique_words)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Compute cosine similarity matrix
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

```python
# Create a mapping from podcast name to index
indices = pd.Series(df.index, index=df['Name']).drop_duplicates()

def recommend_podcasts(title, n=5):
    idx = indices.get(title)
    if idx is None:
        print("Podcast not found.")
        return []

    # Get similarity scores for this podcast
    sim_scores = list(enumerate(cosine_sim[idx]))

    # Sort by similarity and exclude the first (itself)
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)[1:n+1]

    # Get the indices of top recommendations
    podcast_indices = [i[0] for i in sim_scores]

    # Return the top recommended podcasts
    return df[['Name', 'Podcast URL']].iloc[podcast_indices]
```

```python
recommend_podcasts("The Daily", n=5)
```
